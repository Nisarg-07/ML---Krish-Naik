```python
import matplotlib.pyplot as plt
```

```python
# Line plot
x = [1,2,3,4,5]
y = [5,4,3,2,1]
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line chart')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]
 
# Customized line plot 
plt.plot(x,y, color = 'black',linestyle= '--', marker='o',linewidth = 2,markersize = 7)
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line chart')
plt.show()
```

```python
# Multiple plots

## Sample data

x = [1,2,3,4,5]
y = [1,4,8,12,16]
y1 = [1,2,3,4,5]

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(x,y,color='blue',linestyle ='--')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(y,y1,color= 'black',linestyle='--')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(x,y1,color= 'Red',linestyle='--')
plt.title('Plot 3')
plt.show()
```

```python
# Bar plot 
categories = ['A','B','C','D','E']
values = [10,20,30,40,50]

plt.bar(categories,values,color = 'pink')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title('Bar graph')
plt.show()
```

```python
# Histogram 

data = [1,1,1,2,3,4,4,3,2,5,5,2,4,3,2,1,2,3,3,5,4]

plt.hist(data, bins=5,color = 'red',edgecolor='black')
```

```python
# Scatter plot 

x = [1,2,3,4,5,6]
y = [3,4,5,6,7,8]

plt.scatter(x,y,color = 'black', marker='*')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')
plt.show()
```

```python
# Pie chart

labels = ['A','B','C','D']
values = [20,30,10,40]
color = ['yellow','red','pink','blue']
explode = [0,0,.2,0]

plt.pie(values,labels=labels,colors= color,autopct='%1.1f%%',shadow=True,explode=explode)

```

```python
# viz on Sales data
import pandas as pd
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Sales by product category

total_sales_by_product = df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product
```

```python
# Bar graph
total_sales_by_product.plot(kind='bar',color='lightblue')
```

```python
# sales trend over the time 
sales_trend = df.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
