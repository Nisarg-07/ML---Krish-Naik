```python
import seaborn as sns
```

```python
# Basic plotting 
```

```python
tips_df= sns.load_dataset('tips')
tips_df.head()
```

```python
tips_df.columns
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips_df)
plt.title('Scatter plot')
plt.show()
```

```python
# Line plot
sns.lineplot(x = 'size',y= 'total_bill',data = tips_df)
plt.title('Line plot of size by total bill')
plt.show()
```

```python
# Categorical plots

# Bar plot 

sns.barplot(x = 'day',y='total_bill',data = tips_df)
plt.title('Bar graph days by total bill')
plt.show()
```

```python
# Box plot 
sns.boxplot(x = 'day',y='total_bill',data = tips_df)
plt.title('Boxplot')
plt.show()
```

```python
# Violin plot 
sns.violinplot(x = 'day',y='total_bill',data = tips_df)
plt.title('violin plot')
plt.show()
```

```python
# Histogram 
sns.histplot(tips_df['total_bill'],bins=10, kde=True)
```

```python
# KDE plot 
sns.kdeplot(tips_df['total_bill'],fill=True)
```

```python
# Pairplot 
sns.pairplot(tips_df)
```

```python
tips_df.columns
```

```python
# Heatmap 
corr = tips_df[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
sales_df.columns
```

```python
#Bar plot :- total sales by product

plt.figure(figsize=(10,7))
sns.barplot(x = 'Product Category', y = 'Total Revenue', data = sales_df, estimator=sum)
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Bar plot of total revenue by product category')
plt.show()
```

```python
#Bar plot :- total sales by region

plt.figure(figsize=(10,7))
sns.barplot(x = 'Region', y = 'Total Revenue', data = sales_df, estimator=sum)
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.title('Bar plot of total revenue by Region')
plt.show()
```
